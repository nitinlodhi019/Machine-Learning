#Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Upload Data

```python
df = pd.read_csv('/content/student_habits_performance.csv')
```

#Data understanding

```python
print(df.shape)
df.head()
```

```python
#data info
df.info()
```

```python
#missing data
df.isnull().sum()
```

```python
#data statistical
df.describe()
```

#Fill missing value

```python
#fill missing value with mode
df['parental_education_level'].fillna(df['parental_education_level'].mode()[0], inplace=True)
```

#Drop irrelevant column

```python
#drop student_id column
df.drop(['student_id'],axis = 1,inplace = True)
```

#Data Visualization

```python
#correlation heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Box over exam_score or parental_education_level and gender
plt.figure(figsize = (10,6))
plt.rcParams.update({'font.size': 12})
plt.title('outliers')
sns.boxplot(x = 'parental_education_level', y = 'exam_score',data = df,hue = 'gender')
plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

```python
#Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=df, hue = 'gender', palette='Set2')
plt.title('Gender Distribution')
plt.show()
```

```python
#Exam Score distributin
plt.figure(figsize=(8, 5))
sns.histplot(df['exam_score'], bins=15, kde=True, color='salmon')
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python
#Study Vs Exam score
plt.figure(figsize=(8, 5))
sns.scatterplot(x='study_hours_per_day', y='exam_score', hue='gender', data=df)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours Per Day')
plt.ylabel('Exam Score')
plt.show()
```

```python
#Exam Score by Diet Quality
plt.figure(figsize=(8, 5))
sns.boxplot(x='diet_quality', y='exam_score', data=df, palette='Set3', hue = 'diet_quality')
plt.title('Exam Score by Diet Quality')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='mental_health_rating', y='exam_score', data=df)
plt.title('Exam Score by Mental Health Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x='social_media_hours', y='exam_score', data=df, color='coral')
plt.title("Social Media Hours vs Exam Score")

plt.subplot(1, 2, 2)
sns.scatterplot(x='netflix_hours', y='exam_score', data=df, color='slateblue')
plt.title("Netflix Hours vs Exam Score")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='attendance_percentage', y='exam_score', hue='part_time_job', data=df)
plt.title("Attendance vs Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='internet_quality', y='mental_health_rating', data=df)
plt.title("Mental Health Rating by Internet Quality")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='parental_education_level', y='exam_score', data=df)
plt.xticks(rotation=45)
plt.title("Parental Education Level vs Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='extracurricular_participation', y='exam_score', data=df, palette="Pastel1", hue = 'extracurricular_participation')
plt.title("Exam Score Distribution by Extracurricular Participation")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='study_hours_per_day', y='exam_score', hue='sleep_hours', size='sleep_hours', data=df, palette='cool', sizes=(20, 200))
plt.title("Study Hours and Sleep Hours on Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='diet_quality', y='exam_score', hue='mental_health_rating', data=df)
plt.title("Exam Score by Diet and Mental Health")
plt.legend(title='Mental Health Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='exam_score', hue='part_time_job', data=df)
plt.title("Exam Score by Gender and Part-Time Job Status")
plt.show()
```

```python
df['performance_level'] = pd.qcut(df['exam_score'], q=4, labels=['Low', 'Medium', 'High', 'Top'])

plt.figure(figsize=(8, 5))
sns.countplot(x='performance_level', hue='extracurricular_participation', data=df)
plt.title("Performance Levels vs Extracurricular Participation")
plt.show()
```
