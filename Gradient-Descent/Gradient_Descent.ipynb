```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y =make_regression(n_samples=10, n_features=1 ,n_informative=1, n_targets=1, noise=80, random_state=10)
```

```python
print(X.shape)
X
```

```python
print(y.shape)
y
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression

reg =LinearRegression()
reg.fit(X,y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red')
```

```python
# lets apply GD asssuming slope is constant m = 80.28395105
# and let's assume the starting value for intercept b = 0
y_pred = ((80.28395105 * X) + 0).reshape(10)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label ='OLS')
plt.plot(X, y_pred, color='green', label='b = 0')
plt.legend()
plt.show()
```

```python
# we take m = constant (reg.coef_)

m = reg.coef_
b = 100
```

```python
loss_slope = -2 * np.sum(y - m*X.ravel() - b )
loss_slope
```

```python
#lets take learning rate = 0.1
lr = 0.01
step_size = lr * loss_slope

#calculate new intercept
b = b - step_size
b
```

```python
y_pred1 = ((m * X) + b).reshape(10)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label = 'OLS')
plt.plot(X, y_pred1, color ='#A3E4D7', label='b = {}'.format(b))
plt.plot(X, y_pred, color='green', label = 'b = 0')
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(10)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='green',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
