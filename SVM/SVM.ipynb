#SVM
 A support vector machine (SVM) is a supervised machine learning algorithm that classifies data by finding an optimal line or hyperplane that maximizes the distance between each class in an N-dimensional space.

##NonLinear Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
```

```python
from sklearn.datasets import make_circles, make_moons, make_blobs
X, y = make_circles(n_samples=100, factor=.1, noise=.1)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

### Linear Kernal

```python
classifier = SVC(kernel = 'linear')
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
#zero_one_colourmap = ListedColormap(('green', 'blue'))  --old version
zero_one_colourmap = ListedColormap([(0, 1, 0), (0, 0, 1)])

def plot_decision_boundary(X, y, clf):
  X_set, y_set = X, y
  X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1,
                                 stop = X_set[:, 0].max() + 1,
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1,
                                 stop = X_set[:, 1].max() + 1,
                                 step = 0.01))
  plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
               alpha = 0.75, cmap = zero_one_colourmap)
  plt.xlim(X1.min(), X1.max())
  plt.ylim(X2.min(), X2.max())

  for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = zero_one_colourmap(i), label = j)

  plt.title('Decision Boundary')
  plt.xlabel('X1')
  plt.ylabel('X2')
  plt.legend()
  return plt.show()
```

```python
plot_decision_boundary(X, y, classifier)
```

### Nonlinear kernal (RBF)

f(x) = e^(-x^2) -- change the dimension of dataset




The radial basis function (RBF) kernel is a popular kernel function used in support vector machines (SVMs) for classification and regression tasks.


**What it is**

The RBF kernel is a function that measures how similar two samples are. It's a generalized form of kernelization and is often used in kernelized learning algorithms.


**How it works**

The RBF kernel computes the similarity between two samples, represented as feature vectors, by measuring their Euclidean distance. The RBF kernel's value decreases with distance, ranging from zero at infinite distance to one when the samples are identical.


**Why it's popular**

The RBF kernel is popular because it's similar to the Gaussian distribution. It's also the default kernel used in the scikit-learn library.


**How it's used**

The RBF kernel is a non-parametric model that works well with high-dimensional and non-linear data

```python
rbf_kernal = SVC(kernel = 'rbf')
rbf_kernal.fit(X_train, y_train)
```

```python
y_pred = rbf_kernal.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
plot_decision_boundary(X, y, rbf_kernal)
```

### Nonlinear kernal(poly)
