# multi classification metrics

##Iris

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
iris = pd.read_csv('/content/iris.csv')
iris
```

```python
iris.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
iris['species'] = encoder.fit_transform(iris['species'])
```

```python
iris.sample(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,0:-1], iris.iloc[:,-1],test_size=0.2, random_state=2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('Accuracy of LR: ', accuracy_score(y_test, y_pred1))
print('Accuracy of DT: ', accuracy_score(y_test, y_pred2))
```

```python
print('Logistic Regression confusion matrix\n')
pd.DataFrame(confusion_matrix(y_test, y_pred1), columns=list(range(0,3)))
```

```python
print('Decision Tree confusion matrix\n')
pd.DataFrame(confusion_matrix(y_test, y_pred2), columns=list(range(0,3)))
```

```python
result = pd.DataFrame()
result['Actual label'] = y_test
result['LR Prediction'] = y_pred1
result['DT prediction'] = y_pred2
result.sample(5)
```

```python
from sklearn.metrics import precision_score,recall_score
precision_score(y_test,y_pred1,average=None)
```

```python
recall_score(y_test,y_pred1,average=None)
```

## Digit-recognizer

```python
digit = pd.read_csv('/content/train.csv')
digit.sample(5)
```

```python
digit.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digit.iloc[:,1:], digit.iloc[:,0], test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pred1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pred2))
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred1),columns=list(range(0,10)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pred2),columns=list(range(0,10)))
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred1,average='weighted')
```

```python
recall_score(y_test,y_pred1,average='weighted')
```

```python
f1_score(y_test,y_pred1,average='weighted')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
