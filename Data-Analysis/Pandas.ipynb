```python
def run_scored(batsman_name):
  vk=delivery[delivery['batsman']==batsman_name]
  return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head().index[0]
```

# Isin function

```python
# find the most destructive death over batsman
# with the help of SR
# min batsman 200 balls in 16-20
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
# we founded all batsman whos play 200+ balls
# now find runs scored by all these batsman
# balls played by these batsman
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```
